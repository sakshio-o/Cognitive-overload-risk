## 8. GUI Demonstration

A lightweight GUI was implemented using Tkinter.

Users input:

- Total Clicks
- Avg Clicks
- Click Std
- Avg Delay

The system outputs:

- Overload Risk Probability
- Predicted Class (0 or 1)

This simulates an early-warning academic monitoring tool.

```python
import customtkinter as ctk
import numpy as np
import joblib

# Load trained model (same directory)
model = joblib.load("overload_model.pkl")

ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

app = ctk.CTk()
app.title("Cognitive Overload Risk Predictor")
app.geometry("520x600")
app.resizable(False, False)

def predict():
    try:
        values = np.array([[
            float(entry_total.get()),
            float(entry_avg.get()),
            float(entry_std.get()),
            float(entry_delay.get())
        ]])
        
        prob = float(model.predict_proba(values)[0][1])

        # Risk categorization based on probability
        if prob < 0.3:
            level = "LOW RISK"
            color = "#4CAF50"
            conclusion = "Student shows stable engagement patterns."
        elif prob < 0.7:
            level = "MODERATE RISK"
            color = "#FFA726"
            conclusion = "Behavior suggests potential cognitive strain."
        else:
            level = "HIGH RISK"
            color = "#FF4C4C"
            conclusion = "Strong overload indicators detected. Intervention recommended."

        # Update UI
        risk_label.configure(text=level, text_color=color)
        probability_label.configure(text=f"Overload Probability: {prob:.2%}")
        conclusion_label.configure(text=conclusion)
        progress.set(prob)

    except ValueError:
        risk_label.configure(text="INPUT ERROR", text_color="#FF4C4C")
        probability_label.configure(text="Enter valid numeric values.")
        conclusion_label.configure(text="")
        progress.set(0)


# Title
title_label = ctk.CTkLabel(
    app,
    text="AI-Based Student Overload Detector",
    font=("Segoe UI", 20, "bold")
)
title_label.pack(pady=20)

# Input Card
card = ctk.CTkFrame(app, corner_radius=15)
card.pack(padx=30, pady=10, fill="both")

def create_input(label_text):
    label = ctk.CTkLabel(card, text=label_text)
    label.pack(pady=(12,4))
    entry = ctk.CTkEntry(card, width=260)
    entry.pack()
    return entry

entry_total = create_input("Total Clicks")
entry_avg = create_input("Average Clicks")
entry_std = create_input("Click Standard Deviation")
entry_delay = create_input("Average Submission Delay")

predict_button = ctk.CTkButton(
    card,
    text="Analyze Student",
    command=predict,
    width=220,
    height=45,
    corner_radius=25
)
predict_button.pack(pady=20)

# Progress Bar
progress = ctk.CTkProgressBar(card, width=320)
progress.set(0)
progress.pack(pady=10)

# Output Section
output_frame = ctk.CTkFrame(app, corner_radius=15)
output_frame.pack(padx=30, pady=20, fill="both")

risk_label = ctk.CTkLabel(
    output_frame,
    text="",
    font=("Segoe UI", 18, "bold")
)
risk_label.pack(pady=(15,5))

probability_label = ctk.CTkLabel(
    output_frame,
    text="",
    font=("Segoe UI", 14)
)
probability_label.pack(pady=5)

conclusion_label = ctk.CTkLabel(
    output_frame,
    text="",
    font=("Segoe UI", 13),
    wraplength=450
)
conclusion_label.pack(pady=(5,15))

app.mainloop()
```
