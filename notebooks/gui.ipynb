## 8. GUI Demonstration

A lightweight GUI was implemented using Tkinter.

Users input:

- Total Clicks
- Avg Clicks
- Click Std
- Avg Delay

The system outputs:

- Overload Risk Probability
- Predicted Class (0 or 1)

This simulates an early-warning academic monitoring tool.

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import joblib

model = joblib.load("overload_model.pkl")

def predict():
    try:
        values = np.array([[
            float(entry_total.get()),
            float(entry_avg.get()),
            float(entry_std.get()),
            float(entry_delay.get())
        ]])
        
        prob = model.predict_proba(values)[0][1]
        pred = model.predict(values)[0]
        
        if pred == 1:
            interpretation = "⚠ High Overload Risk"
        else:
            interpretation = "✓ Normal Engagement"
        
        result_text.set(
            f"{interpretation}\nProbability: {prob:.2f}"
        )
        
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numeric values.")

root = tk.Tk()
root.title("Student Overload Predictor")
root.geometry("350x300")

tk.Label(root, text="Total Clicks").grid(row=0, column=0, padx=10, pady=5)
entry_total = tk.Entry(root)
entry_total.grid(row=0, column=1)

tk.Label(root, text="Avg Clicks").grid(row=1, column=0, padx=10, pady=5)
entry_avg = tk.Entry(root)
entry_avg.grid(row=1, column=1)

tk.Label(root, text="Click Std").grid(row=2, column=0, padx=10, pady=5)
entry_std = tk.Entry(root)
entry_std.grid(row=2, column=1)

tk.Label(root, text="Avg Delay").grid(row=3, column=0, padx=10, pady=5)
entry_delay = tk.Entry(root)
entry_delay.grid(row=3, column=1)

tk.Button(root, text="Predict", command=predict).grid(row=4, column=0, columnspan=2, pady=15)

result_text = tk.StringVar()
tk.Label(root, textvariable=result_text, fg="blue").grid(row=5, column=0, columnspan=2)

root.mainloop()
```
